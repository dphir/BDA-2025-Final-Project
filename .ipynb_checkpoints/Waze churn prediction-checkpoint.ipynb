# üè¢ **1. Business Context**

Waze, a community-driven navigation app, faces a critical challenge in retaining its user base amid increasing competition in the mapping and mobility space. User churn‚Äîdefined as the discontinuation of app usage over a defined period‚Äîdirectly impacts Waze‚Äôs revenue streams (e.g., advertising, partnerships), market share, and long-term growth. While Waze collects extensive behavioural and usage data, the lack of actionable insights into why and when users churn limits the effectiveness of retention strategies.

> **Business Context Highlight**
> 
> - **Critical Challenge**: User churn threatens revenue, market share, and long-term growth.  
> - **Current Gap**: Despite extensive behavioral and usage data, Waze lacks actionable insights to predict and mitigate churn effectively.


# ‚ùó **2. Problem Statement**

This project aims to address the problem of user churn in the Waze app. Understanding why users leave is important for maintaining customer loyalty and developing effective retention strategies. Churn prediction is critical for Waze‚Äôs business operations, helping to forecast user behaviour and optimize engagement efforts. This problem is relevant because reducing churn directly impacts user base stability and business growth.

> **Problem statment Conclusion**

> - No clear understanding of **why** and **when** users leave, limiting retention strategy effectiveness.  


# üéØ **3. Project Goals (Objectives)**  
This project aims to develop a churn prediction model for Waze based on user behavioral and usage data, such as driving days, kilometers driven per day, and device type. The objectives are to predict user churn by identifying key features to:

* Accurately identify at-risk users.

## **Overview of Data Quality Issues**

### This section validate the logic about the data based on common sense


````python
# 1. Check if any numeric columns contain negative values
numeric_columns = [
    'sessions', 'drives', 'total_sessions', 'n_days_after_onboarding',
    'total_navigations_fav1', 'total_navigations_fav2',
    'driven_km_drives', 'duration_minutes_drives',
    'activity_days', 'driving_days'
]

print("Check for negative values:")
print(" ")
for col in numeric_columns:
    if (df[col] < 0).any():
        print(f"{col} has negative values: {(df[col] < 0).sum()} rows")
    else:
        print(f"{col} has no negative values")

# 2. Logical consistency: drives should not exceed sessions
invalid_drives = (df['drives'] > df['sessions']).sum()

print(" ")
print("Logical consistency:")
print(f"\n Drives > Sessions: {invalid_drives} rows" if invalid_drives > 0 else "\n All drives are <= sessions")

# 3. Logical consistency: driving_days should not exceed activity_days
invalid_driving_days = (df['driving_days'] > df['activity_days']).sum()
print(f" Driving Days > Activity Days: {invalid_driving_days} rows" if invalid_driving_days > 0 else " All driving_days <= activity_days")

# 4. Logical consistency: activity_days should not exceed n_days_after_onboarding
invalid_activity_days = (df['activity_days'] > df['n_days_after_onboarding']).sum()
print(f" Activity Days > Days After Onboarding: {invalid_activity_days} rows" if invalid_activity_days > 0 else " All activity_days <= n_days_after_onboarding")

# 5. Validate device values (should be a small set like Android, iPhone)
print("\nDevice values:")
print(df['device'].value_counts())

# 6. Invalid categorical values
print("\nUnique device types:")
print(df['device'].value_counts())
````

````python
def get_issues(row):
    issues = []

    if row['duration_minutes_drives'] > 10000:
        issues.append('duration_too_long')
    if pd.notnull(row['driving_days']) and pd.notnull(row['activity_days']) and row['driving_days'] > row['activity_days']:
        issues.append('driving_days_exceed_activity_days')
    if row['n_days_after_onboarding'] > 4000:
        issues.append('onboarding_days_too_high')
    if pd.notnull(row['drives']) and row['drives'] > 0:
        km_per_drive = row['driven_km_drives'] / row['drives'] if pd.notnull(row['driven_km_drives']) else np.nan
        if km_per_drive < 1 or km_per_drive > 1000:
            issues.append('km_per_drive_unrealistic')
    if pd.notnull(row['driving_days']) and row['driving_days'] > 0:
        avg_duration = row['duration_minutes_drives'] / row['driving_days'] if pd.notnull(row['duration_minutes_drives']) else np.nan
        if avg_duration > 720:
            issues.append('avg_drive_duration_per_day_too_high')
    if row['driving_days'] == 0:
        issues.append('never_drives')

    return issues if issues else np.nan

# Apply to a fresh copy
df_check = df.copy()

# Convert numeric
cols = ['drives', 'driving_days', 'driven_km_drives', 'duration_minutes_drives', 'activity_days', 'n_days_after_onboarding']
df_check[cols] = df_check[cols].apply(pd.to_numeric, errors='coerce')

# Apply validation
df_check['validation_issues'] = df_check.apply(get_issues, axis=1)

# Optionally: count how many issues per row
df_check['n_issues'] = df_check['validation_issues'].apply(lambda x: len(x) if isinstance(x, list) else 0)

# Display only rows with issues
df_check[df_check['n_issues'] > 0].head()
````

## **Device Analysis (Android vs iOS)**

````python
sns.countplot(data=waze_data, x='device', hue='label')
plt.title("Churn by Device Type")
plt.show()

device_churn = waze_data.groupby('device')['label'].value_counts().unstack()

# Normalize by device to get proportions
device_churn_ratio = device_churn.div(device_churn.sum(axis=1), axis=0)

# Plot pie chart per device
fig, axes = plt.subplots(1, len(device_churn_ratio), figsize=(10, 8))
for i, device in enumerate(device_churn_ratio.index):
    axes[i].pie(device_churn_ratio.loc[device].values, 
                labels=device_churn_ratio.columns,
                autopct='%1.1f%%',
                startangle=90,
                textprops={'fontsize': 10})
    axes[i].set_title(f'Device: {device}')
    
fig.suptitle('Churn Distribution by Device')
plt.tight_layout()
plt.show()
````

## **Customer Segmentation by Engagement: High vs Low Engagement Users**

````python
# Create binary engagement level based on median of sessions
median_sessions = waze_data['sessions'].median()
waze_data['engagement_level'] = waze_data['sessions'].apply(lambda x: 'High' if x >= median_sessions else 'Low')

# Countplot
sns.countplot(x='engagement_level', hue='label', data=waze_data)
plt.title("Churn by Engagement Level")
plt.xlabel("Engagement Level")
plt.ylabel("Count")
plt.show()

# Group and normalize
engagement_churn = waze_data.groupby('engagement_level')['label'].value_counts().unstack()
engagement_churn_ratio = engagement_churn.div(engagement_churn.sum(axis=1), axis=0)

# Pie charts for each engagement level
fig, axes = plt.subplots(1, len(engagement_churn_ratio), figsize=(10, 8))
for i, level in enumerate(engagement_churn_ratio.index):
    axes[i].pie(engagement_churn_ratio.loc[level].values, 
                labels=engagement_churn_ratio.columns,
                autopct='%1.1f%%',
                startangle=90,
                textprops={'fontsize': 10})
    axes[i].set_title(f'Engagement Level: {level}')  #Fixed title

fig.suptitle('Churn Distribution by Engagement Level', fontsize=12)
plt.tight_layout()
plt.show()
````

## **Outliers and Distributions**

````python

# Column sets
engagement_cols = ['sessions', 'drives', 'activity_days', 'driving_days',
                   'driven_km_drives', 'duration_minutes_drives']
fav_cols = ['total_navigations_fav1', 'total_navigations_fav2']
all_cols = engagement_cols + fav_cols

# Subplot settings
n_cols = 2
n_rows = int(np.ceil(len(all_cols) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4))
fig.suptitle('Skewed Distributions', fontsize=16)

for i, col in enumerate(all_cols):
    row = i // n_cols
    col_idx = i % n_cols
    ax = axes[row, col_idx] if n_rows > 1 else axes[col_idx]

    sns.histplot(waze_data[col], kde=True, ax=ax, bins=30, color='salmon')

    # Calculate statistics
    median_val = waze_data[col].median()
    mean_val = waze_data[col].mean()

    # Plot lines
    ax.axvline(median_val, color='red', linestyle='--', label='Median')
    ax.axvline(mean_val, color='green', linestyle='-.', label='Mean')

    # Annotate values
    ax.text(median_val, ax.get_ylim()[1]*0.9, f'Median: {median_val:.2f}', 
            color='red', ha='right', fontsize=6, rotation=0)
    ax.text(mean_val, ax.get_ylim()[1]*0.9, f'Mean: {mean_val:.2f}', 
            color='green', ha='left', fontsize=6, rotation=0)

    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel('')
    ax.set_ylabel('Frequency')
    ax.legend()

# Hide unused subplots
for j in range(i + 1, n_rows * n_cols):
    row = j // n_cols
    col_idx = j % n_cols
    ax = axes[row, col_idx] if n_rows > 1 else axes[col_idx]
    ax.axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
````

**There are 3 commonly used methods to deal with outliers.**

1. **Dropping the outliers**

   The simplest and easiest way to handle outliers is by removing them entirely, compared to the other two methods. However, this approach reduces the dataset size and narrows the data dimensions. Removing outliers can be a good choice if you have a large dataset because the risk of losing valuable information is small and it likely won‚Äôt significantly affect your model‚Äôs performance. But keep in mind, removing outliers means deleting all records with extreme values, which can be problematic if your data is sensitive, prone to multicollinearity, or limited in size, as it might negatively impact your model results.

2. **Winsorize method**

   Winsorization limits the outliers by capping the extreme values instead of removing them completely. This is done by replacing extreme values with less extreme percentile values. Unlike the arithmetic mean, winsorized data limits the influence of outliers that can distort the overall results, rather than treating all data points equally. For example, in a 5% winsorization, the lowest 5% of data points are replaced by the 5th percentile value, and the highest 5% are replaced by the 95th percentile value. This method helps maintain the overall structure of the dataset while reducing the effect of outliers, making it a robust alternative for datasets with extreme values.

3. **Log transformation**

    Log transformation reduces data skewness and helps normalize the distribution of data. However, its effectiveness depends on the data itself ‚Äî applying a log transform may not always reduce skewness and could even worsen it if not suitable. Therefore, it is important to analyze the data distribution before applying this method.

**In this case (predicting customer churning in Waze app, we consider to apply **Winsorization**, Why Winsorization is suitable for this case?**

* **Presence of real extreme outliers:**
    Features like driven_km_drives with very high extreme values (e.g., 21,183 km) indicate a distinct group of users (super drivers). Removing these outliers completely would discard important information.

* **Highly right-skewed distributions:**
    Winsorization effectively trims the tails of the distribution, reducing the impact of outliers without deleting data. This helps maintain the overall data structure and allows models to learn more stable patterns.

* **Models plan to use:**
    * Logistic Regression (logreg) is very sensitive to outliers, so applying Winsorization helps stabilize the inputs.
    * Random Forest and XGBoost are generally more robust to outliers, but cleaned data often improves performance and interpretability.

we do not choose Log transformation, eventhough some of our data are positive and highly skewed, because one or two features are zero, so it doesn‚Äôt fit with our needs. In addition, removing outliers entirely might be too extreme and lose valuable user information.

Anyway, before we implement winsorize; we need to define data range, usually use percentile between 5%

### **What Is Encoding in Machine Learning?**

In machine learning, encoding is the process of converting categorical data into numerical format so that algorithms can work with it. Based on diagram above we can see that there are different type of encoding techniques to help us during feature engineering process. Here, I will explain about Nominal and Ordinal Encoding which realted to our project. 
So, both **Nominal and ordinal encoding** is a commonly use for feature engineering process and clasify as **classical encoders**. 

#### **1. Ordinal Encoding**
This method is used when the categories have a meaningful order (e.g., rating levels like ‚Äúlow,‚Äù ‚Äúmedium,‚Äù ‚Äúhigh‚Äù). Two popular techniques include **Label Encoder** and **Target-guided Encoder**. the diferent of both is about how they encode the data and the similarity is related to label data. if Label Encoding assigns a unique integer to each category. For example, ‚ÄúApartment‚Äù becomes 0, ‚ÄúVilla‚Äù becomes 1, and ‚ÄúTownhouse‚Äù becomes 2. However, Target-Guided Encoding encodes categories based on their relationship with the target variable.
If the mean house prices are $210,000 for apartments, $525,000 for villas, and $305,000 for townhouses, these values are used for encoding. This reflects the relationship with the target but can risk data leakage if not done carefully.

#### a. Label Encoding
- Assigns a **unique integer** to each category.
- Example:  
  `"Apartment" ‚Üí 0`, `"Townhouse" ‚Üí 1`, `"Villa" ‚Üí 2`
- ‚úÖ Simple and efficient.
- ‚ùå Can mislead linear models due to the assumption of order if not present.

**Use When:**
- Categories are ordered.
- Using models like **decision trees**, **XGBoost**, etc.

#### b. Target-Guided Encoding (e.g., Mean Encoding)
- Replaces each category with the **mean of the target variable**.
- Example:  
  `"Apartment" ‚Üí $210,000`, `"Townhouse" ‚Üí $305,000`, `"Villa" ‚Üí $525,000`

‚úÖ Captures relationship with the target.  
‚ö†Ô∏è **High risk of data leakage** if not done with cross-validation.

**Use When:**
- You have ordinal or nominal data.
- Strong correlation between category and target.
- Using **tree-based models**.


#### **2. Nominal Encoding**
For non-ordinal data where categories don‚Äôt have a natural order (like colors or cities), nominal encoders are appropriate:
* **One-Hot Encoder**: Creates binary columns for each category. Works well when there are few categories.
* **Dummy Encoder:** Similar to one-hot encoding but drops one category to avoid multicollinearity.
* **Mean Encoder:** Replaces each category with the mean of the target variable for that category.

#### a. One-Hot Encoding
- Creates a **binary column** for each category.
- Example:  
  `"Blue"` ‚Üí `[0, 1, 0]`, `"Green"` ‚Üí `[0, 0, 1]`, `"Red"` ‚Üí `[1, 0, 0]`

‚úÖ Good for small number of categories.  
‚ùå Can cause high dimensionality with many unique values.

**Use When:**
- Categories are unordered.
- Suitable for **linear models**, **neural networks**, etc.

#### b. Dummy Encoding
- Similar to one-hot, but **drops one column** to avoid multicollinearity.
- Useful for **linear regression** models.

#### c. Mean Encoding (Again)
- Also applicable for nominal data.
- Encodes based on **target mean per category**.

‚úÖ Efficient for high-cardinality data.  
‚ö†Ô∏è Risk of overfitting and leakage if not handled carefully.

#### **Summary: Similarities, Differences & When to Use**

| **Technique**           | **Uses Target (y)?** | **For Ordered Categories?** | **High Cardinality?** | **Risk of Leakage** | **Recommended Models**          |
|-------------------------|----------------------|------------------------------|------------------------|---------------------|----------------------------------|
| **Label Encoding**      | No                   | ‚úÖ Yes                       | ‚úÖ Yes                | No                  | Tree-based models                |
| **One-Hot Encoding**    | No                   | ‚ùå No                        | ‚ùå No                 | No                  | Linear, distance-based, NN       |
| **Dummy Encoding**      | No                   | ‚ùå No                        | ‚ùå No                 | No                  | Linear regression                |
| **Target/Mean Encoding**| ‚úÖ Yes               | ‚úÖ/‚ùå Both                   | ‚úÖ Yes                | ‚ö†Ô∏è Yes              | Tree-based models                |

### Handling Imbalanced Data in Machine Learning

<div style="text-align: center;">
  <img src="imbalance.png" alt="resampling" width="500" />
</div>

**Reference:**
* https://machinelearningmastery.com/5-effective-ways-to-handle-imbalanced-data-in-machine-learning/
* https://www.linkedin.com/posts/datascience-trainer_datascience-machinelearning-ai-activity-7224380712978669569-KxUK


Imbalanced datasets are common in real-world machine learning problems, where one class significantly outnumbers the other. This can lead to models that are biased toward the majority class and perform poorly in detecting the minority class ‚Äî which is often the class of interest (e.g., fraud detection, churn prediction, etc.).

There are several effective techniques to handle imbalanced data:

#### 1. Resampling Techniques

- **Random Oversampling**  
  Increases the number of minority class examples by duplicating them randomly. This can help balance the dataset but risks **overfitting** because the model sees repeated instances.

- **Random Undersampling**  
  Reduces the number of majority class examples by randomly removing them. While this also balances the dataset, it may lead to **information loss** if important examples are discarded.

- **SMOTE (Synthetic Minority Oversampling Technique)**  
  Instead of duplicating, SMOTE creates synthetic samples by interpolating between existing minority instances. This is more effective than random oversampling, as it **reduces overfitting** and maintains more information diversity.

#### 2. Algorithm-Level Approaches

- **Cost-Sensitive Learning**  
  Assigns a higher penalty (cost) for misclassifying the minority class. Many classifiers (like decision trees and SVM) can incorporate class weights to emphasize the minority class.

#### 3. Ensemble Methods

- **Balanced Bagging or Boosting**  
  Combines undersampling or oversampling with ensemble models like Random Forest or XGBoost. Techniques like **Balanced Random Forest** or **EasyEnsemble** apply sampling in each bootstrap iteration, maintaining model robustness while handling imbalance.

#### 4. Anomaly Detection Models

- When the minority class is extremely rare (e.g., <1%), treating the problem as anomaly detection can be more effective than standard classification.

#### 5. Evaluation Metrics for Imbalanced Data

- Use appropriate metrics like **F1-score**, **Precision-Recall AUC**, or **ROC AUC**, instead of accuracy ‚Äî which can be misleading in imbalanced datasets.

#### **Why We Use SMOTE?**

We observed that our dataset is imbalanced, with around **80% of instances labeled as non-churned (retained)**. To prevent our models from being biased toward the majority class, we apply **SMOTE**.

Unlike random oversampling, **SMOTE creates synthetic samples** from the minority class. This helps to:

- Avoid overfitting caused by repeated samples,
- Improve generalization,
- Preserve more useful patterns in the data.

our dataset is imbalanced, with about 80% of instances labeled as non-churned (retained). Handling this imbalance is crucial to avoid biased model performance towards the majority class.

thus, We choose SMOTE over simple resampling methods because SMOTE creates synthetic minority class samples instead of just duplicating existing ones. This reduces the risk of overfitting and introduces less bias compared to random oversampling, which can replicate the same minority instances and thus bias the model.

### Logistic Regression defaults to a 0.5 threshold for classification, but in churn prediction, lowering this threshold can improve recall for the churn class.

````python
from sklearn.metrics import precision_recall_curve, f1_score, confusion_matrix, classification_report

# Train the model
logreg.fit(X_train_logreg_sm, y_train_logreg_sm)
y_proba = logreg.predict_proba(X_test_logreg)[:, 1]


precision, recall, thresholds = precision_recall_curve(y_test_logreg, y_proba)

plt.plot(thresholds, precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision and Recall vs Threshold (default - 0.5) -Logistic Regression')
plt.legend()
plt.grid()
plt.show()

# choose threshold manually
threshold = 0.35 
y_pred_threshold = (y_proba >= threshold).astype(int)


# Evaluate the model at this threshold
print(f"Classification report at threshold = {threshold}:\n")
print(classification_report(y_test_logreg, y_pred_threshold))

print(f"Confusion Matrix:\n{confusion_matrix(y_test_logreg, y_pred_threshold)}")

# comparation to find best threshold Compute and plot F1-score vs threshold to find the best threshold
f1_scores = [f1_score(y_test_logreg, (y_proba >= t).astype(int)) for t in thresholds]

plt.figure(figsize=(8, 6))
plt.plot(thresholds, f1_scores)
plt.xlabel('Threshold')
plt.ylabel('F1 Score')
plt.title('F1 Score vs Threshold -Logistic Regression')
plt.grid()
plt.show()


````

## **What are SHAP Values?**

SHAP (SHapley Additive exPlanations) values are a way to explain the output of any machine learning model. It uses a game theoretic approach that measures each player's contribution to the final outcome. In machine learning, each feature is assigned an importance value representing its contribution to the model's output.

SHAP values show how each feature affects each final prediction, the significance of each feature compared to others, and the model's reliance on the interaction between features.

## **SHAP Implementation - Random Forest**

````python
import shap

# Random Forest model for SHAP
rf_model = RandomForestClassifier(random_state=42)

# Training model with data handled by SMOTE 
rf_model.fit(X_train_tree_sm, y_train_tree_sm)

# Sampling data test
X_sample = X_test_tree.sample(200, random_state=42)

# Create SHAP explainer for Random Forest Model
explainer_rf = shap.TreeExplainer(rf_model, approximate=True)

# Calculate SHAP values for sampled test set - Random Forest
shap_values_rf = explainer_rf.shap_values(X_sample)
````

````python
# shap_values_rf has shape (samples, features, classes)

# Select the SHAP values for class 1 (positive class)
shap_values_class1 = shap_values_rf[:, :, 1]

# Plot summary SHAP for class 1
shap.summary_plot(shap_values_class1, X_sample, feature_names=X_sample.columns.tolist())
````

## **SHAP Implementation - XGBoost**

## **What is A/B Testing?**

A/B testing is one of the most popular controlled experiments used to optimize web marketing strategies. A/B testing is an experiment where you show two variants (A and B) to different segments of users at the same time and analyze which variant achieves a desired outcome better.

<div style="text-align: center;">
  <img src="https://miro.medium.com/v2/resize:fit:640/format:webp/1*NMguspN33VZOQoXhauIwpw.png" alt="encode" width="200"/>
</div>

**Key Concepts of A/B testing:**

- Variant A: This is usually the control ‚Äì the current version.
- Variant B: This is the treatment ‚Äì the new version you're testing.

Metric/Goal: The outcome you want to improve (e.g., click-through rate, conversion, signups, purchases).

**How It Works**

1. Choose a goal for example to increase email signup rate.
2. Create two versions
    * Version A: Existing webpage.
    * Version B: New webpage with a different CTA (Call to Action) button color.

3. Split the audience randomly (50% see A, 50% see B).
4. Collect data on user behavior to track who signed up, clicked, bought, etc.
5. Analyze the results by Using statistical methods to determine which version performed better with significance.

````python
retention_rates = low_engagement_users.groupby('group')['retained_flag'].mean()
print(retention_rates)
````

### **Perform hypothesis testing (proportion z-test)**

Using proportion z-test to compare retention rates.

````python
from statsmodels.stats.proportion import proportions_ztest

# Count of retained users in each group
success_counts = low_engagement_users.groupby('group')['retained_flag'].sum().values

# Total users in each group
nobs = low_engagement_users.groupby('group')['retained_flag'].count().values

# Perform z-test
stat, pval = proportions_ztest(success_counts, nobs)

print(f"Z-statistic: {stat:.4f}")
print(f"P-value: {pval:.4f}")

if pval < 0.05:
    print("Result: Statistically significant difference between groups.")
else:
    print("Result: No statistically significant difference between groups.")
````
